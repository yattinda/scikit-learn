```python
import pandas as pd
import numpy as np
import cv2
import csv
import matplotlib.pyplot as plt
```

```python
city1_image= cv2.imread('./data/city1.jpg')
city2_image= cv2.imread('./data/city2.jpg')
forest1_image = cv2.imread('./data/forest1.jpg')
forest2_image = cv2.imread('./data/forest2.jpg')

coordinate = pd.read_csv('./data/coordinate.csv', header = None)
coordinate = coordinate.values

bgr = []
for i in range(0, 750, 10):
    bgr.append(np.append(city1_image[coordinate[i][1], coordinate[i][0], :], 1))

for i in range(0, 750, 10):
    bgr.append(np.append(city2_image[coordinate[i][1], coordinate[i][0], :], 1))
    
for i in range(0, 750, 10):
    bgr.append(np.append(forest1_image[coordinate[i][1], coordinate[i][0], :], 0))
    
for i in range(0, 750, 10):
    bgr.append(np.append(forest2_image[coordinate[i][1], coordinate[i][0], :], 0))

with open('./data/bgr.csv', 'w') as file:
    writer = csv.writer(file, lineterminator='\n')
    writer.writerows(bgr)
```

```python
train = pd.read_csv('./data//bgr.csv', header = None)
train = train.values

B = train[:,0]
G = train[:,1]
R = train[:,2]
```

```python
B_0 = B[:150]
B_1 = B[150:]
R_0 = R[:150]
R_1 = R[150:]
plt.scatter(R_0,B_0)
plt.scatter(R_1,B_1, c="red")
plt.xlabel('R')
plt.ylabel('B')
```

```python
G_0 = G[:150]
G_1 = G[150:]
R_0 = R[:150]
R_1 = R[150:]
plt.scatter(R_0,G_0)
plt.scatter(R_1,G_1, c="red")
plt.xlabel('R')
plt.ylabel('G')
```

```python
G_0 = G[:150]
G_1 = G[150:]
B_0 = B[:150]
B_1 = B[150:]
plt.scatter(B_0,G_0)
plt.scatter(B_1,G_1, c="red")
plt.xlabel('G')
plt.ylabel('B')
```

```python
bgr_array = cv2.imread('./data/test1.jpg')
answer = pd.read_csv('./data/test1_coord.csv', header = None)
answer = answer.values
bgr = []
for i in range(10):
    bgr.append(np.append(bgr_array[answer[i][1], answer[i][0], :], 0))
for i in range(10,20):
    bgr.append(np.append(bgr_array[answer[i][1], answer[i][0], :], 1))

with open('./data/answer1_bgr.csv', 'w') as file:
    writer = csv.writer(file, lineterminator='\n')
    writer.writerows(bgr)
```
